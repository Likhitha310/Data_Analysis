# WebScraping

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Academy_Award%E2%80%93winning_films"
# Send a GET request to the URL
req = requests.get(url)
```

```python
req
```

```python
soup = BeautifulSoup(req.content)
```

```python
soup.prettify()
```

```python
arr = [ ]
for i in soup.findAll('tr'):
    arr.append(i)
```

```python
arr[0]
```

```python
arr[1]
```

```python
arr = [ ]
for i in soup.findAll('td'):
    arr.append(i)
```

```python
arr[0]
```

```python

# Define the regular expression pattern to remove HTML tags
pattern = r'<.*?>'

# List to store cleaned data
cleaned_data = []

# Iterate over each item in the dataset
for item in arr:
    # Convert the BeautifulSoup Tag object to a string
    html_string = str(item)
    
    # Use re.sub() to remove HTML tags
    cleaned_string = re.sub(pattern, "", html_string)
    
    # Append the cleaned string to the cleaned data list
    cleaned_data.append(cleaned_string)

# Print or further process the cleaned data
for cleaned_item in cleaned_data:
    print(cleaned_item)
```

```python
Film = []
Year = []
Awards = []
Nominations = []
count = 0
for i in soup.findAll('td'):
    if count == 0:
        Film.append(i)
        count += 1
    elif count == 1:
        Year.append(i)
        count += 1
    elif count == 2:
        Awards.append(i)
        count += 1
    else:
        count = 0
        Nominations.append(i)
```
